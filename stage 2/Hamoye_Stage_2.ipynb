```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.linear_model import LinearRegression
```

```python
scaler = MinMaxScaler()
```

```python
df.columns
```

```python
feat = ['Appliances', 'T_kitchen', 'H_kitchen', 'T_living',
       'H_living', 'T_laundry', 'H_laundry', 'T_office', 'H_office',
       'T_bathroom', 'H_bathroom', 'T_outside_north', 'H_outside_north',
       'T_ironing_room', 'H_ironing_room', 'T_teenager_room',
       'H_teenager_room', 'T_parent_room', 'H_parent_room', 'T_out',
       'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility', 'Tdewpoint', 'rv1',
       'rv2']
```

```python
df1 = pd.DataFrame(scaler.fit_transform(df[feat]), columns=feat)
```

```python
df1.head()
```

```python
X = df1.drop(columns=['Appliances'])
```

```python
y = df1['Appliances']
```

```python
X.columns
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, y_train.shape  
```

```python
X_test.shape, y_test.shape
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
lin_reg_pred = lin_reg.predict(X_test)
```

```python
print(f'mse : {mean_squared_error(y_test, lin_reg_pred)}')
print(f'rmse : {mean_squared_error(y_test, lin_reg_pred, squared=False)}')
print(f'mae : {mean_absolute_error(y_test, lin_reg_pred)}')
print(f'r2_score : {r2_score(y_test, lin_reg_pred)}')
print(f'rss : {np.sum(np.square(y_test - lin_reg_pred))}')
```

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)
```

```python
print(f'mse : {mean_squared_error(y_test, lasso_pred)}')
print(f'rmse : {mean_squared_error(y_test, lasso_pred, squared=False)}')
print(f'mae : {mean_absolute_error(y_test, lasso_pred)}')
print(f'r2_score : {r2_score(y_test, lasso_pred)}')
print(f'rss : {np.sum(np.square(y_test - lasso_pred))}')
```

```python
ridge = Ridge(alpha=0.4)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)
```

```python
print(f'mse : {mean_squared_error(y_test, ridge_pred)}')
print(f'rmse : {mean_squared_error(y_test, ridge_pred, squared=False)}')
print(f'mae : {mean_absolute_error(y_test, ridge_pred)}')
print(f'r2_score : {r2_score(y_test, ridge_pred)}')
print(f'rss : {np.sum(np.square(y_test - ridge_pred))}')
```

```python
el = ElasticNet(alpha=0.01)
el.fit(X_train, y_train)
el_pred = el.predict(X_test)
```

```python
def get_weights(model, feat, col_name):
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```python
Linear_model_weight = get_weights(lin_reg, X_train, 'Linear_model_weight')
lasso_model_weight = get_weights(lasso, X_train, 'lasso_model_weight')
ridge_model_weight = get_weights(ridge, X_train, 'ridge_model_weight')
elastic_model_weight = get_weights(el, X_train, 'elastic_model_weight')
```

```python
final_weights = pd.merge(Linear_model_weight, lasso_model_weight, on='Features')
final_weights = pd.merge(final_weights, ridge_model_weight, on='Features')
final_weights = pd.merge(final_weights, elastic_model_weight, on='Features')

final_weights
```
